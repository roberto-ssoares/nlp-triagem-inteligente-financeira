<font size="4" color='#CC403E'><strong>📌 1. Contexto de Negócio</strong></font>

<font size="2" color='#66666'>

>- Instituições financeiras que operam concessão de crédito convivem com um desafio central:
    - *ampliar sua carteira de clientes* e
    - *gerar receita de forma sustentável*,
    - *sem comprometer a qualidade do portfólio*.

>- No contexto deste projeto, o problema é *inspirado em uma fintech que concede crédito* para públicos com:
    - *pouca informação histórica*,
    - *cenário comum em operações de inclusão financeira*.
    - *a medida que a carteira cresce, também cresce a preocupação com a inadimplência e com a necessidade de decisões mais robustas e escaláveis*.

>- Nesse tipo de operação, depender apenas de análises descritivas ou critérios fixos de negócio pode não ser suficiente.
    - Torna-se necessário desenvolver mecanismos analíticos capazes de estimar o risco individual de cada cliente e apoiar:
        - decisões de concessão,
        - revisão de política e
        - monitoramento da carteira.

>- Assim, este projeto propõe a construção de um modelo preditivo de risco de inadimplência,
    - com foco em apoiar decisões relacionadas à concessão de crédito em cartão.

---

</font></div>

<div style="background-color:#f3f2ee">
    
<font size="4" color='#CC403E'><strong>📌 2. Contexto do case / motivação institucional</strong></font>

<font size="2" color='#66666'>

>- A motivação deste projeto está alinhada a um cenário clássico de risco de crédito:
    - uma instituição financeira observa o crescimento da inadimplência no mercado
    - e passa a **demandar uma solução preditiva que permita antecipar o risco de default**.

>- No desafio de negócio que inspira este trabalho, a necessidade é clara:
    - ***construir um modelo capaz de gerar um score de risco para apoiar a contratação de produtos de crédito***.

>- A proposta não é apenas prever inadimplência,
    - mas ***transformar essa previsão em um insumo de decisão utilizável por áreas de negócio, risco e planejamento***.

>- Portanto, este projeto busca responder a uma necessidade real do setor financeiro:
    - ***equilibrar crescimento da carteira***,
    - ***controle de perdas***
    - e ***ampliação do acesso ao crédito com maior inteligência analítica***.

---

</font></div>

<div style="background-color:#f3f2ee">
    
<font size="4" color='#CC403E'><strong>📌 3. Definição do problema de negócio</strong></font>

<font size="2" color='#66666'>

>- O problema de negócio pode ser resumido da seguinte forma:

>- <font color = 'red'>***Como identificar, no momento da análise, quais clientes apresentam maior probabilidade de inadimplência, de modo a apoiar decisões mais seguras e eficientes na concessão de crédito?***</font>

>- Essa questão é relevante porque decisões de crédito envolvem trade-offs importantes:
    - *aprovar clientes com alto risco pode elevar perdas financeiras*;
    - *rejeitar clientes com bom potencial pode reduzir receita e restringir crescimento*;
    - *políticas pouco calibradas podem comprometer competitividade e sustentabilidade da operação*.

>- Dessa forma, o projeto busca oferecer uma abordagem quantitativa para priorizar risco,
    - melhorar a consistência das decisões
    - e apoiar a construção de políticas de concessão mais aderentes ao perfil dos clientes.

---

</font></div>

<div style="background-color:#f3f2ee">
    
<font size="4" color='#CC403E'><strong>📌 4. Problema Analítico</strong></font>

<font size="2" color='#66666'>

>- **Sob a perspectiva analítica, o desafio consiste em construir um modelo supervisionado de classificação binária capaz de estimar a probabilidade de inadimplência de um cliente**, com base em:
    - variáveis cadastrais,
    - comportamentais,
    - financeiras
    - e históricas.

>- Em termos práticos, o modelo deverá aprender padrões nos dados que permitam diferenciar clientes com maior e menor risco de default.

>- **A saída esperada do modelo não é apenas uma classe binária, mas preferencialmente uma probabilidade de inadimplência, que possa posteriormente ser transformada em**:
    - ***score de risco***;
    - ***ranking de clientes***;
    - ***faixas de risco***;
    - ***apoio a políticas de aprovação, reprovação ou revisão manual***.

---

</font></div>

<div style="background-color:#f3f2ee">
    
<font size="4" color='#CC403E'><strong>📌 5. Perguntas Analíticas</strong></font>

<font size="2" color='#66666'>

>- Pergunta principal:
    - **Quais clientes apresentam maior probabilidade de inadimplência em cartão de crédito?**

>- Perguntas secundárias:
    - *Quais variáveis possuem maior relação com o risco de default?*
    - *Um modelo simples e interpretável, como regressão logística, já entrega performance satisfatória?*
    - *Modelos baseados em árvores e ensembles conseguem capturar padrões adicionais relevantes?*
    - *Como transformar probabilidades previstas em uma regra de decisão utilizável pelo negócio?*
    - *Qual trade-off entre capturar maus pagadores e evitar rejeição excessiva de bons clientes?*

---

</font></div>

<div style="background-color:#f3f2ee">
    
<font size="4" color='#CC403E'><strong>📌 6. Objetivo Geral</strong></font>

<font size="3" color='#66666'>

>- ***Desenvolver um modelo preditivo capaz de estimar a probabilidade de inadimplência de clientes de cartão de crédito,***
    - ***apoiando decisões mais inteligentes em:***
        - ***concessão,***
        - ***priorização de risco***
        - ***e definição de políticas analíticas.***

---

</font></div>

<div style="background-color:#f3f2ee">
    
<font size="4" color='#CC403E'><strong>📌 7. Objetivos Específicos</strong></font>

<font size="3" color='#66666'>

>- 💡 Para atingir o objetivo geral, este projeto buscará:
    - 1. *compreender o contexto de negócio e o papel da modelagem de risco na concessão de crédito;*
    - 2. *explorar e diagnosticar a base de dados, identificando qualidade, padrões e limitações;*
    - 3. *preparar os dados para modelagem, com tratamento de inconsistências e transformação de variáveis;*
    - 4. *desenvolver features relevantes para risco de inadimplência;*
    - 5. *construir um baseline interpretável com regressão logística;*
    - 6. *comparar o baseline com modelos mais avançados, como Random Forest e XGBoost;*
    - 7. *avaliar desempenho por métricas estatísticas e métricas orientadas ao negócio;*
    - 8. *simular uma lógica inicial de decisão baseada em score e threshold;*
    - 9. *documentar resultados, limitações e próximos passos.*

---

</font></div>

<div style="background-color:#f3f2ee">
    
<font size="4" color='#CC403E'><strong>📌 8. Hipóteses Iniciais</strong></font>

<font size="3" color='#66666'>

>- Antes de analisar os dados, algumas **hipóteses de negócio podem orientar a investigação**:
    - *clientes com histórico de atraso ou pagamento irregular tendem a apresentar maior risco de default;*
    - *maior comprometimento de renda ou maior utilização de limite pode estar associado a maior probabilidade de inadimplência;*
    - *instabilidade financeira e histórico de crédito insuficiente podem aumentar incerteza na análise;*
    - *variáveis comportamentais e históricas tendem a ser mais informativas do que variáveis puramente cadastrais;*
    - *modelos de ensemble podem capturar interações e não linearidades que um baseline linear não captura.*

>- Essas hipóteses serão verificadas, refinadas ou rejeitadas nas etapas seguintes do projeto.

---

</font></div>

<div style="background-color:#f3f2ee">
    
<font size="4" color='#CC403E'><strong>📌 9. Valor Esperado para o Negócio</strong></font>

<font size="3" color='#66666'>

- A construção de um modelo de inadimplência pode gerar valor em diferentes frentes:

>- **Redução de perdas**
    - *Ao identificar clientes com maior propensão ao default, a empresa pode reduzir exposição a operações de maior risco.*

>- **Melhoria da concessão**
    - *A decisão deixa de depender exclusivamente de regras fixas ou análises manuais, ganhando consistência, escala e rastreabilidade.*

>- **Priorização de análise**
    - *Casos com risco intermediário podem ser encaminhados para revisão manual, enquanto extremos de risco podem ser tratados por política automática.*

>- **Segmentação da carteira**
    - *O score gerado pelo modelo pode ser usado para classificar clientes em faixas de risco e apoiar estratégias comerciais, limites e monitoramento.*

>- **Eficiência operacional**
    - *Modelos analíticos podem reduzir tempo de decisão e aumentar produtividade das equipes de crédito e risco.*

<font size="4" color='#CC403E'><strong>📌 10. Critérios de Sucesso</strong></font>

<font size="2" color='#66666'>

- O sucesso deste projeto deve ser avaliado em duas dimensões complementares: **estatística e negócio**.

>- **Critérios estatísticos**
    - boa **capacidade discriminatória** entre adimplentes e inadimplentes;
    - **desempenho consistente** em treino, validação e teste;
    - **estabilidade mínima** entre amostras;
    - **comparação clara** entre baseline e modelos avançados.

>- **Critérios de negócio**
    - **capacidade de ordenar** clientes por risco;
    - **utilidade do score** para apoiar decisão;
    - **possibilidade de definição** de thresholds operacionais;
    - **clareza sobre o custo relativo** de falsos positivos e falsos negativos.

>- **Métricas prioritárias**
    - Para este projeto, serão consideradas especialmente relevantes:
        - **ROC-AUC**
        - **Gini**
        - **KS**
        - **Precision**
        - **Recall**
        - **F1-score**
        - **matriz de confusão**
        - **análise por faixas/ranking de score**

<font size="4" color='#CC403E'><strong>📌 11. Impacto dos Erros de Classificação</strong></font>

<font size="2" color='#66666'>

- Em problemas de risco de crédito, os erros do modelo não têm o mesmo peso.

>- **Falso negativo**
    - O modelo classifica como baixo risco um cliente que se tornará inadimplente.
>- ***Impacto: potencial perda financeira, aumento da inadimplência e deterioração da carteira.***

>- **Falso positivo**
    - O modelo classifica como alto risco um cliente que pagaria corretamente.
>- ***Impacto: perda de oportunidade comercial, possível redução de receita e restrição excessiva de crédito.***

>- Assim, a calibragem do modelo não deve buscar apenas “acurácia”, mas sim equilíbrio entre captura de risco, eficiência comercial e apetite de risco da instituição.

<font size="4" color='#CC403E'><strong>📌 13. Premissas e Limitações Iniciais</strong></font>
